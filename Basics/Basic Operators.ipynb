#Logical Operators

And: and

Or: or

Not: not

And

Both conditions must be true for the and condition to be true

Take variables x and z below

We have x = 1.0 and z = 2.0

Four Possible Cases
1. x == 1.0 and z == 2.0
2. x != 1.0 and z == 2.0
3. x == 1.0 and z != 2.0
4. x != 1.0 and z != 2.0

With the "and" logical operator, only one of these will evaluate to true.

You will see that below when you compile the code.

For and to result in true, both inputs must be true.

```python
x = 1.0
z = 2.0

if(x == 1.0 and z == 2.0):
  print("Both conditions are satified, AND!")

if(x != 1.0 and z == 2.0):
  print("Both conditions are satified, AND!")

if(x == 1.0 and z != 2.0):
  print("Both conditions are satified, AND!")

if(x != 1.0 and z != 2.0):
  print("Both conditions are satified, AND!")
```

Or

At least one of the conditions must be true for the or condition to be true

```python
myBool1 = True
myBool2 = False

if(myBool1 or myBool2):
  print("Boolean 1: True Boolean2: False")

if(myBool1 or myBool1):
  print("Boolean 1: True Boolean2: True")

if(myBool2 or myBool1):
  print("Boolean 1: False Boolean2: True")

if(myBool2 or myBool2):
  print("Boolean 1: False Boolean2: False")     #will NOT print since both are false
```

image.png

Or Truth Table Explanation

A = Input 1

B = Input 2

Q = Result of A and B

0 = False

1 = True

See how when Q = 1, we have 3 cases:
1. A = 1, B = 0
2. A = 0, B = 1
3. A = 1, B = 1

As long as A or B has a 1 (is true), then Q (or) will result to true



Not

Negates the current logical statement

not(True) --> false

not(False) --> true

```python
myBoolean1 = True

print(not(myBoolean1))
```

```python
myBoolean1 = False
print(not(myBoolean1))
```
