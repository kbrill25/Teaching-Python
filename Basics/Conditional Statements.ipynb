#Concept of Conditional Statements

Code is executed if a certain condition is satisfied.

For example, if it is hot outside, I will drink water. But if it is cold outside, I will drink coffee. Otherwise, I will drink nothing.

First Statement: If it is hot outside, I will drink water.

Condition: hot outside

Action: drink water


Second Statement: But if it is cold outside, I will drink coffee.

Since we are at the second statement, we know it is not hot outside. But is it cold outside?

Condition: cold outside

Action: drink coffee


Third Statement: Otherwise, I will drink nothing.

The first and second statements were NOT satisfied. So now, we do nothing.

Condition: N/A

Action: do nothing 


Note: when programming the condition can be anything that we want
In order for the action to occur, the condition must be satisfied (true)

#If statement

Idea: If a certain statement is true, then we do something
See code block below for syntax and basic examples(this code will NOT compile)

If a certain statement is false, then we do NOTHING and the action does not occur.

The condition is represented is parentheses
Do Something represents some sort of action 

#Syntax

```python
if(statement == True):
  do something
```

```python
if(statement1 == False):
  do something
```

```python
if(statement2 == 5 or statement3 > 2):
  do something
  do more stuff
```

We can have if statements within if statements, these are called nested if-statements

```python
if(statement4 == 2 and statement1 == statement2):
  do something
  do more stuff

  if(statement5 == "abc"):
    do even more stuff
```

#Else-if statement

IMPORTANT: they must always be preceded by an if-statement

Think of like an "otherwise if this condition is true, we do XYZ"

If an if statement is true, then we do not check an else-if statement


#Syntax

```python
if(statement == True):
  do something

elif(statement == False):
  do something
```

```python
if(statement == True):
  do something

  if(statement1 > statement2):
    do something
  elif(statement2 > statement1):
    do something
  elif(statement1==statement2):
    do something

elif(statement == False):
  do something

  if(statement1 > statement2):
    do something
  elif(statement2 > statement1):
    do something
  elif(statement1==statement2):
    do something
```

#Else Statement

No condition parentheses

The action is completed if the preceding statements do not evaluate to true

Must be preceded by an if or else-if statement

#Syntax

```python
if(statement1 == True):
  do something

else:
  do something
```

```python
if(statement1 == True):
  do something

if(statement2 == True):
  do something

else:
  do something
```

```python
if(statement1 == True):
  do something

elif(statement2 == True):
  do something

else:
  do something
```

Assume that we are only evaluating statement 1: If the temperature is hot, then we drink water.

```python
temperature = "hot"
```

```python
if(temperature == "hot"):  #condition  
  print("drink water")      #action
```

Now's let's make this a little more complicated by adding in the second statement, if the temperature is cold, then we drink coffee.

```python
temperature = "cold"
```

```python
if(temperature == "hot"):  #condition  
  print("drink water")      #action

elif(temperature == "cold"):  #condition
  print("drink coffee")
```

What happened in the code above?

1. First, the compiler compares the temperature variable to "hot". 
2. Since they are unequal, the compiler "moves on" to the next statement since the first statement is false. 
3. Now, the compiler is checking to see if temperature is equal to "cold", which evaluates to true. 
4. Since the statement is true, we move onto the "action", so we print out "drink coffee".
